# TASK-Image to Pencil Sketch with Python

Importing necessary libraries

```python
import cv2
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

Importing the Image

```python
img = cv2.imread("lab.jpeg")
```

Reading the Image and Displaying it 

```python
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
plt.axis(False)
plt.show()
```

Converting Original image to Grayscale

```python
grey_scale_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_scale_image)
```

Inverting the Image

```python
inverted_image = cv2.bitwise_not(grey_scale_image)
plt.imshow(inverted_image)

```

Using Gaussian Function to Blur the Image

```python
blur_image = cv2.GaussianBlur(inverted_image, (111,111), 0)
plt.imshow(blur_image)
```

```python
inverted_blur_image = cv2.bitwise_not(blur_image)
plt.imshow(inverted_blur_image)
```

Inverting the Blur Image

```python
sketch = cv2.divide(grey_scale_image, inverted_blur_image, scale=256.0)
plt.imshow(sketch)
```

```python
cv2.imwrite("Sketch.png", sketch)
```

**Final Sketch Image**

```python
plt.figure(figsize = (8,8))
Final = cv2.cvtColor(sketch, cv2.COLOR_RGB2BGR)
plt.imshow(Final)
plt.axis(False)
plt.show()
```
